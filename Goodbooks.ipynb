```python
mappings = ratings.groupby(by = ['user_id','book_id']).mean()
```

```python
mappings['rating'].reset_index()['user_id'].value_counts().sample(10)
```

```python
minimum_books = 3
```

```python
counts = mappings['rating'].reset_index()['user_id'].value_counts()
guess_from = counts.loc[counts > minimum_books]

guess_from.tail(5)
```

```python
to_guess = counts.loc[counts <= minimum_books]

to_guess.sample(10)
```

```python
reseted_mappings = mappings.reset_index()
reseted_mappings = reseted_mappings.loc[reseted_mappings['user_id'].isin(guess_from.index)]
reseted_mappings = reseted_mappings.groupby(by = ['user_id','book_id']).mean()
reseted_mappings.head(10)
```

```python
reseted_mappings_dict = reseted_mappings['rating'].to_dict()
reseted_mappings.reset_index(inplace = True)
```

```python
book_mappings = {}
reverse_book_mappings = {}
for i,x in enumerate(mappings.reset_index()['book_id'].unique()):
    book_mappings[x] = i
    reverse_book_mappings[i] = x
    
user_mappings = {}
reverse_user_mappings = {}
for i,x in enumerate(mappings.reset_index()['user_id'].unique()):
    user_mappings[x] = i
    reverse_user_mappings[i] = x
    
N_book = mappings.reset_index()['book_id'].nunique()
N_user = mappings.reset_index()['user_id'].nunique()
```

```python
list(reseted_mappings_dict.keys())[0]   , list(reseted_mappings_dict.values())[0]
```

```python
from scipy.sparse import dok_matrix
M = dok_matrix((N_user+1, N_book+1), dtype=np.uint8)

for k in reseted_mappings_dict:
    M[   user_mappings[k[0]]    ,   book_mappings[k[1]]    ] = int(reseted_mappings_dict[k])
```

```python
print(M.sum() )
print(M.sum()/len(reseted_mappings_dict) )
```

```python
matching = 10

from sklearn.neighbors import NearestNeighbors

neighbours = NearestNeighbors(n_neighbors = matching, algorithm = 'brute')
neighbours.fit(M)
```

```python
reseted_mappings_test = mappings.reset_index()
reseted_mappings_test = reseted_mappings_test.loc[reseted_mappings_test['user_id'].isin(to_guess.index)]
reseted_mappings_test = reseted_mappings_test.groupby(by = ['user_id','book_id']).mean()
reseted_mappings_test.head(10)
```

```python
reseted_mappings_test_dict = reseted_mappings_test['rating'].to_dict()
reseted_mappings_test.reset_index(inplace = True)
```

```python
from scipy.sparse import dok_matrix
M_test = dok_matrix((N_user+1, N_book+1), dtype=np.uint8)

for k in reseted_mappings_test_dict:
    M_test[   user_mappings[k[0]]    ,   book_mappings[k[1]]    ] = reseted_mappings_test_dict[k]
```

```python
print(M_test.sum() )
print(M_test.sum()/len(reseted_mappings_test_dict) )
```

```python
col_M_test = M_test.tocsr()
```

```python
total_recommendations = []
i = 0

for k in reseted_mappings_test['user_id'].unique():
    now = col_M_test[   user_mappings[  k  ]  ]
    densed = M  [   neighbours.kneighbors(now, return_distance = False)[0]   ].todense()
    
    recommendation = pd.DataFrame(  densed.sum(0) / ((densed>0).sum(0)+1)  ).T
    
    seen = np.where(now.todense() > 0)[1]
    
    recommendations = recommendation.loc[~recommendation.index.isin(seen)].sort_values(by = 0, ascending = False).iloc[0:15]
    
    recommend_these = [  reverse_book_mappings[x]   for x in   recommendations.index     ]
    
    total_recommendations.append(    (k, recommendations[0].sum(), recommend_these)   )
    
    if i > 100:
        break
    i += 1
```

```python
sorts = sorted(total_recommendations, key = lambda x: (x[1]), reverse = True)[0:4]
sorts
```

```python
select = [x[0] for x in sorts]
select
```
